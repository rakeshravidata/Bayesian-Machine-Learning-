```python
import numpy as np
import pandas as pd
import sklearn.mixture as mx 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats, linalg
```

```python
#initializing means and variances
mu1 = [-2.5, 2.5]
sigma1 = [[0.5, 0.081],[0.081, 0.7]]

mu2 = [-4, -2]
sigma2 = [[0.4, 0.02],[0.02, 0.3]]
#considered 0.02 instead of 0.002

mu3 = [2, -1]
sigma3 = [[0.6, 0.531],[0.531, 0.9]]

mu4 = [0.1, 0.2]
sigma4 = [[0.5, 0.22],[0.22, 0.8]]

mu5 = [3.0, 3.0]
sigma5 = [[0.88, 0.2],[0.2, 0.22]]
```

```python

def plot_results(X, Y_, means, covariances, index, title):
    splot = plt.subplot(2, 1, 1 + index)
    for i, (mean, covar, color) in enumerate(zip(
            means, covariances, color_iter)):
        v, w = linalg.eigh(covar)
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        u = w[0] / linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

        # Plot an ellipse to show the Gaussian component
        angle = np.arctan(u[1] / u[0])
        angle = 180. * angle / np.pi  # convert to degrees
        ell = mpl.patches.Ellipse(mean, v[0], v[1], 180. + angle, color=color)
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(0.5)
        splot.add_artist(ell)

    plt.xlim(-6., 6.)
    plt.ylim(-5., 6.)
    plt.xticks(())
    plt.yticks(())
    plt.title(title)
```

```python
def gen_normal(mu,sigma,cnt):
    x, y = np.random.multivariate_normal(mu, sigma, cnt).T
    n = pd.DataFrame([x,y]).T
    n.columns = ['x','y']
    return n
```

```python
n1 = gen_normal(mu1,sigma1,60)
n2 = gen_normal(mu2,sigma2,60)
n3 = gen_normal(mu3,sigma3,60)
n4 = gen_normal(mu4,sigma4,60)
n5 = gen_normal(mu5,sigma5,60)

plt.plot(n1.x, n1.y, 'x')
plt.axis('equal')
plt.show()
```

```python
ncomb = pd.concat([n1,n2,n3,n4,n5])
print(ncomb.shape)
ncomb1 = ncomb.copy()
```

```python
gm = mx.GaussianMixture(n_components=25,init_params='random',random_state=0,verbose=2)
ncomb['pred'] = gm.fit_predict(ncomb)
ncomb.pred.unique()
```

```python
plot_results(ncomb[['x','y']].values, ncomb['pred'].values, gm.means_, gm.covariances_, 0,  'Expectation Maximization')
```

```python
fig, ax = plt.subplots()
ax.scatter(ncomb['x'], ncomb['y'],c=ncomb['pred'])
sns.despine(); ax.legend()
ax.set(xlabel='X', ylabel='Y', title='Expectation Maximization');
```

```python
bgm = mx.BayesianGaussianMixture(n_components=25,init_params='random',random_state=1,verbose=2)
ncomb1['pred'] = bgm.fit_predict(ncomb1)
```
